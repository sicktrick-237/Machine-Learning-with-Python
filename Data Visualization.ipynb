```python
from matplotlib import pyplot as plt
import seaborn as sb
import pandas as pd


df = pd.read_csv("stocks.csv")
x = df['UR']
y = df['SP']

# SeaBorn Line Plot
lineplot = sb.lineplot(x,y,data=df)
lineplot
```

```python
# MatPlotLib Line Plot
x = x.sort_values()
y = y.sort_values()
graph = plt.plot(x,y)
```

```python
# Seaborn Barplot
sb.barplot(x,y,data=df,palette='copper')
```

```python
# Bar Chart MatPlotLib
plt.bar(x,y)
plt.show()
```

```python
# SeaBorn Scatterplot
sb.scatterplot(x,y,data=df,palette='copper')
```

```python
# Matplotlib scatterplot
plt.scatter(x,y)
```
