```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AgglomerativeClustering as AggCluster

customer_data = pd.read_csv('shopping-data.csv')
data = customer_data.iloc[:,3:5].values  # Here we are considering only the two columns 3 and 4

# print(data)

import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("CUstomer Dendogram")
link = shc.linkage(data,method='single') # How to link data points based on which method
dend = shc.dendrogram(link) # 
# ward divides data based on ward linkage
# affinity = eculidean or manhattan distance


cluster = AggCluster(n_clusters=6, affinity='euclidean')
cluster.fit_predict(data)
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=cluster.labels_,cmap='rainbow')
plt.show()
print(cluster.labels_)
```
